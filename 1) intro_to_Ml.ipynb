# Machine Learning

Machine Learning (ML) is a field of Artificial Intelligence that focuses on building systems that learn patterns from data and make predictions or decisions without being explicitly programmed.



Instead of writing rules manually, we provide data and allow algorithms to discover patterns automatically.


Example:

Predict house prices based on house features such as size, location, and number of rooms.

Goal of Machine Learning:


Learn a function:

f(X) → Y

where:

X = input features  
Y = predicted output

## Concept

Machine Learning is a method that allows computers to learn patterns from data instead of being programmed with fixed rules.

Traditional programming works like this:

Input → Rules → Output

Machine Learning works differently:

Input + Output examples → Learning algorithm → Model

The model studies the data and learns a pattern that can later be used to make predictions.

Example:

Suppose we want to predict the price of a house.

Input features might include:
- house size
- number of rooms
- location

Output:
- house price

By studying many houses and their prices, the machine learning model learns the relationship between these variables and can predict the price of a new house.

## Math Behind Machine Learning

Most machine learning problems try to learn a function.

f(X) → Y

Where:

X = input features  
Y = output prediction

Example:

X = house size  
Y = house price

The goal of the model is to find a function that predicts Y from X as accurately as possible.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

# generate simple data
X = np.linspace(0,10,50)
y = 2.5 * X + 4 + np.random.normal(0,2,50)

plt.scatter(X,y)

plt.title("Example Machine Learning Dataset")
plt.xlabel("Input Feature")
plt.ylabel("Target Value")

plt.show()
```

Explanation:

Each point represents a data example.

The model will try to find the pattern connecting X and Y.

```python
from sklearn.linear_model import LinearRegression

# create a linear regression model

X_model =  X.reshape(-1,1)

model = LinearRegression()
model.fit(X_model,y)

predictions = model.predict(X_model)
```

Explanation:

Each point represents a data example.

The model will try to find the pattern connecting X and Y.

```python
plt.scatter(X,y, label=" Real Data")
plt.plot(X,predictions, color="red", label="Model Predictions")
plt.legend()
plt.title("Model learning the pattern ")
plt.show()
```

Explanation:

Blue points = actual data
Red line = model prediction

The algorithm tries to find the best pattern that fits the data.
