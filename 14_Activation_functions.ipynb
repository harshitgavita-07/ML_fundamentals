# Activation Functions

## Concept

In neural networks, activation functions introduce **non-linearity** into the model.

Without activation functions, a neural network would behave like a simple linear model,
no matter how many layers it has.

Activation functions decide **whether a neuron should be activated or not**.

Pipeline inside a neuron:

Input Features  
→ Weighted Sum  
→ Activation Function  
→ Output

Common activation functions used in neural networks:

Sigmoid  
ReLU  
Tanh  
Softmax

These functions help neural networks learn **complex patterns in data**.

# Math Behind Activation Functions

1. Sigmoid Function

σ(z) = 1 / (1 + e^(-z))

Range: (0,1)

Used for probability outputs.

---

2. Tanh Function

tanh(z) = (e^z − e^-z) / (e^z + e^-z)

Range: (-1,1)

Centered around zero which helps training.

---

3. ReLU (Rectified Linear Unit)

ReLU(z) = max(0,z)

Very simple and computationally efficient.

---

4. Softmax

Softmax(zᵢ) = exp(zᵢ) / Σ exp(zⱼ)

Converts outputs into **probability distribution** for multi-class classification.

```python
import numpy as np
import matplotlib.pyplot as plt

z = np.linspace(-10,10,200)

sigmoid = 1/(1+np.exp(-z))
tanh = np.tanh(z)
relu = np.maximum(0,z)

plt.figure(figsize=(10,6))

plt.plot(z,sigmoid,label="Sigmoid")
plt.plot(z,tanh,label="Tanh")
plt.plot(z,relu,label="ReLU")

plt.title("Activation Functions")
plt.xlabel("Input")
plt.ylabel("Output")

plt.legend()
plt.grid(True)

plt.show()
```

This graph compares the most common activation functions.

```python
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

data = load_iris()

X_train,X_test,y_train,y_test = train_test_split(
    data.data,
    data.target,
    test_size=0.2,
    random_state=42
)

model = MLPClassifier(
    hidden_layer_sizes=(10,),
    activation="relu",
    max_iter=1000
)

model.fit(X_train,y_train)

accuracy = model.score(X_test,y_test)

print("Model Accuracy:", accuracy)
```

# Summary

Activation functions introduce **non-linearity** into neural networks.

Without them, neural networks cannot learn complex relationships.

Important activation functions:

Sigmoid → used for probability outputs  
Tanh → zero-centered activation  
ReLU → most widely used in deep learning  
Softmax → used in multi-class classification

Modern deep learning models rely heavily on activation functions to learn complex patterns.
