# Support Vector Machines (SVM)

## Concept

Support Vector Machines are powerful supervised learning algorithms used for **classification and regression**.

The main idea of SVM is to find the **best boundary that separates different classes of data**.

This boundary is called a **decision boundary**.

Example:

Imagine we want to classify animals using two features:

Weight  
Nose length  

The algorithm will try to draw a line that separates cats from elephants.

But SVM does something special.

Instead of just finding *any* separating line, it finds the line that:

→ Maximizes the distance between the classes.

This distance is called the **margin**.

The points closest to the boundary are called **Support Vectors**.

These points define the position of the decision boundary.

Pipeline:

Features  
→ Decision Boundary  
→ Maximum Margin  
→ Classification

# Math Behind Support Vector Machines

In SVM we try to find a hyperplane:

w · x + b = 0

Where:

w = weight vector  
x = input features  
b = bias  

The goal is to maximize the **margin** between classes.

Margin = distance between decision boundary and closest points.

Optimization problem:

Minimize:

||w||² / 2

Subject to:

yᵢ (w · xᵢ + b) ≥ 1

Where:

yᵢ = class label (-1 or +1)

This ensures that all points are classified correctly while maximizing the margin.

Support vectors are the points that satisfy:

yᵢ (w · xᵢ + b) = 1

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn import datasets
X,y = datasets.make_blobs(n_samples = 100 , centers = 2 , random_state = 0 )

plt.scatter(X[:,0],X[:,1],c=y,cmap="coolwarm")

plt.title("Dataset for SVM classification ")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

This dataset shows two clusters that can be separated by a decision boundary.

```python
from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

X_train , X_test , y_train , y_test = train_test_split(
    X , y , test_size = 0.4, random_state = 42
)

model = svm.SVC(kernel = "linear")
model.fit(X_train,y_train)

predictions = model.predict(X_test)

accuracy = accuracy_score(y_test,predictions)
print("Accuracy: ",accuracy)

```

```python
from sklearn.datasets import make_circles

X, y = make_circles(n_samples=200, noise=0.1)

from sklearn.svm import SVC

model = SVC(kernel="rbf")

model.fit(X, y)

plt.scatter(X[:,0], X[:,1], c=y, cmap="coolwarm")
plt.title("Non-linear Dataset for Kernel SVM")
plt.show()
```

This example demonstrates non-linear classification using kernels.

# Summary

Support Vector Machines are powerful classification algorithms.

Key ideas:

Decision boundary separates classes.

Margin is maximized to improve generalization.

Support vectors are the closest points to the boundary.

SVM can handle non-linear data using **kernel functions**.

Advantages:

Works well in high dimensional data.

Robust against overfitting.

Limitations:

Slow on very large datasets.

Kernel selection can be tricky.
