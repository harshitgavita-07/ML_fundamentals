# Decision Trees

## Concept

A Decision Tree is a machine learning algorithm that makes decisions
by asking a sequence of simple questions.

It works like a flowchart.

Example:

Is age > 40?
  
   ├── Yes → High risk
   
   └── No → Low risk

Each question splits the dataset into smaller groups.

The goal of the algorithm is to split the data so that
each final group contains mostly one class.

These final groups are called **leaf nodes**.

Decision Trees are easy to understand because they mimic
human decision-making.

Pipeline:

Input Features
→ Decision Questions
→ Data Splitting
→ Leaf Node Prediction

# Math Behind Decision Trees

Decision Trees choose splits using a concept called **Entropy**.

Entropy measures the randomness in the data.

Formula:

Entropy = − Σ p(x) log₂ p(x)

Where:

p(x) = probability of class x

Entropy values:

0 → pure dataset (only one class)
1 → maximum randomness

The algorithm selects splits using **Information Gain**.

Information Gain:

IG = Entropy(parent) − Weighted Entropy(children)

The split with the **highest information gain** is chosen.

This creates better class separation.

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs

X, y = make_blobs(n_samples=200, centers=2, random_state=42)

plt.scatter(X[:,0], X[:,1], c=y, cmap="coolwarm")
plt.title("Dataset for Decision Tree")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

This visualization shows how a dataset can be split into regions.

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split

from sklearn.datasets import load_iris

data = load_iris()

X_train, X_test, y_train, y_test = train_test_split(
    data.data,
    data.target,
    test_size=0.2,
    random_state=42
)

model = DecisionTreeClassifier()

model.fit(X_train, y_train)

predictions = model.predict(X_test)

print(predictions[:10])
```

using this algorithm for training a dataset there

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, predictions)

print("Model Accuracy:", accuracy)
```

```python
from sklearn.tree import plot_tree

plt.figure(figsize=(12,8))

plot_tree(model,
          filled=True,
          feature_names=data.feature_names)

plt.show()
```

This cell visualizes the entire decision tree structure.

# Summary

Decision Trees are interpretable machine learning models
that split data using simple decision rules.

Key concepts:

Tree structure  
Entropy measures randomness  
Information Gain selects best split  

Advantages:

Easy to understand
Works with numerical and categorical data
Requires little preprocessing

Limitations:

Prone to overfitting

Sensitive to small changes in data

Decision Trees become extremely powerful when combined
with ensemble methods such as Random Forest and Boosting.
