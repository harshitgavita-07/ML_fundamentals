# Dimensionality Reduction — Principal Component Analysis (PCA)

## Concept

In many machine learning problems, datasets contain a large number of features.

Example:

Image dataset
→ each pixel becomes a feature

Customer dataset
→ age, income, spending, location, purchase history

Too many features can cause problems:

• slower training
• redundant information
• overfitting

Dimensionality Reduction solves this problem by reducing the number of features
while keeping most of the important information.

Principal Component Analysis (PCA) is the most common dimensionality
reduction technique.

Idea:

Instead of using the original features, PCA creates new features called
**principal components** that capture the most variance in the data.

Pipeline:

High-dimensional data
→ compute variance directions
→ select top principal components
→ reduced feature space

# Math Behind PCA

PCA finds directions where the data varies the most.

Steps:

1️⃣ Standardize the data

2️⃣ Compute covariance matrix

Cov(X) = (1/n) XᵀX

3️⃣ Compute eigenvectors and eigenvalues

Eigenvectors → directions of maximum variance  
Eigenvalues → amount of variance captured

4️⃣ Sort eigenvectors by eigenvalues

5️⃣ Select top K principal components

6️⃣ Project data onto these components

New representation:

Z = XW

Where:

X = original data  
W = principal component vectors

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

data = load_iris()

X = data.data

plt.scatter(X[:,0], X[:,1])

plt.title("Original Feature Space")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")

plt.show()
```

This shows the original feature space before dimensionality reduction.

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)

X_reduced = pca.fit_transform(X)

print("Original shape:", X.shape)
print("Reduced shape:", X_reduced.shape)
```

This reduces the dataset from 4 dimensions --> 2 dimensions

```python
plt.scatter(
    X_reduced[:,0],
    X_reduced[:,1],
    c=data.target,
    cmap="viridis"
)

plt.title("PCA Reduced Dataset")

plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")

plt.show()
```

Now the dataset is visualized in a lower dimensional space.

# Summary

PCA is a dimensionality reduction technique that transforms
high-dimensional data into a smaller number of components.

Key ideas:

Principal components capture maximum variance.

Eigenvectors determine new feature directions.

Eigenvalues determine importance of components.

Advantages:

Reduces computational cost
Removes redundant features
Improves visualization

Limitations:

Principal components may lose interpretability.
