# Optimization Algorithms

## Concept

Optimization algorithms are methods used to update model parameters
so that the loss function decreases during training.

The goal of training is to find the **best weights** that minimize the loss.

Basic training pipeline:

Input Data  
→ Model Prediction  
→ Compute Loss  
→ Compute Gradient  
→ Update Weights  

The simplest optimization algorithm is **Gradient Descent**.

However, modern deep learning uses more advanced optimizers to
train faster and handle complex datasets.

Common optimizers include:

Gradient Descent  
Stochastic Gradient Descent (SGD)  
Momentum  
RMSProp  
Adam

These optimizers control how weights are updated during training.

# Math Behind Optimization Algorithms

Basic Gradient Descent update rule:

w = w − α ∂L/∂w

Where:

w = model weights  
α = learning rate  
∂L/∂w = gradient of loss  

---

Stochastic Gradient Descent (SGD)

Instead of computing gradient on the entire dataset,
SGD updates weights using one sample at a time.

---

Momentum

Momentum adds velocity to updates:

v = βv + α∂L/∂w

w = w − v

This helps the optimizer move faster in the correct direction.

---

Adam Optimizer

Adam combines momentum and adaptive learning rates.

It keeps track of:

First moment (mean of gradients)  
Second moment (variance of gradients)

Adam is widely used in deep learning and LLM training.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-3,3,100)

loss = x**2

plt.plot(x,loss)

plt.title("Loss Landscape Example")
plt.xlabel("Weight Value")
plt.ylabel("Loss")

plt.grid(True)
plt.show()
```

This shows a loss landscape , where optimizers try to move toward the minimum

```python
import numpy as np

# simple quadratic loss
def loss(w):
    return w**2

# gradient
def grad(w):
    return 2*w

w = 5
lr = 0.1

for step in range(20):

    g = grad(w)

    w = w - lr * g

    print("Step:",step,"Weight:",w)
```

This demonstrates  how gradient descent moves weights towards the minimum

```python
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

data = load_iris()

X_train,X_test,y_train,y_test = train_test_split(
    data.data,
    data.target,
    test_size=0.2,
    random_state=42
)

model = MLPClassifier(
    hidden_layer_sizes=(30,),
    solver="adam",
    max_iter=2000
)

model.fit(X_train,y_train)

accuracy = model.score(X_test,y_test)

print("Model Accuracy:", accuracy)
```

Here  the neural networks uses adam optimizer internally .
