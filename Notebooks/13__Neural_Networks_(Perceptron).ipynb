# Neural Networks — Perceptron

## Concept

A Neural Network is a machine learning model inspired by the human brain.

The smallest unit of a neural network is called a **neuron**.

The simplest neural network model is the **Perceptron**.

The perceptron takes input features, multiplies them by weights,
adds a bias, and passes the result through an activation function.

Pipeline:

Input Features  
→ Weighted Sum  
→ Activation Function  
→ Output

Example:

If we want to classify emails as spam or not spam,
the perceptron can learn weights for words like:

"free"
"offer"
"discount"

and produce a classification decision.

The perceptron is essentially a **linear classifier**, but it forms the foundation
of modern neural networks.

# Math Behind a Perceptron

The perceptron computes a weighted sum:

z = w₁x₁ + w₂x₂ + ... + b

Where:

w = weights  
x = input features  
b = bias

The result is passed through an activation function.

Example activation:

Step Function

f(z) = 1 if z ≥ 0  
f(z) = 0 if z < 0

Modern neural networks use smoother activation functions such as:

Sigmoid  
ReLU  
Tanh

Training uses gradient descent to update weights:

w = w − α ∂L/∂w

```python
import numpy as np
import matplotlib.pyplot as plt

X = np.array([[2,3],[3,4],[5,1],[7,3]])
y = np.array([0,0,1,1])

plt.scatter(X[:,0],X[:,1],c=y,cmap="coolwarm")
plt.title("Dataset for Perceptron Classification")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")

plt.show()
```

This visualization shows two classes that the perceptron will try to separate.

```python
import numpy as np

X = np.array([[2,3],[3,4],[5,1],[7,3]])
y = np.array([0,0,1,1])

w = np.zeros(2)
b = 0
lr = 0.1

for epoch in range(100):

    for i in range(len(X)):

        z = np.dot(X[i],w) + b
        prediction = 1 if z >= 0 else 0

        error = y[i] - prediction

        w += lr * error * X[i]
        b += lr * error

print("Weights:", w)
print("Bias:", b)
```

This is a minimal perceptron implementation.

```python
from sklearn.linear_model import Perceptron
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris

data = load_iris()

X_train, X_test, y_train, y_test = train_test_split(
    data.data,
    data.target,
    test_size=0.2,
    random_state=42
)

model = Perceptron()

model.fit(X_train,y_train)

accuracy = model.score(X_test,y_test)

print("Model Accuracy:", accuracy)
```

# Summary

The Perceptron is the simplest neural network model.

Key concepts:

Weighted sum of inputs
Bias term
Activation function

Advantages:

Simple
Foundation of neural networks

Limitations:

Can only solve linearly separable problems.

Modern neural networks solve this by using:

Multiple layers
Nonlinear activation functions
Backpropagation
