## Concept

Supervised Learning is a type of machine learning where the model learns from **labeled data**.

This means each training example contains:

Input features (X)  
Output label (Y)

Example:

Predict house price.

Input features (X):

- house size
- number of rooms
- location

Output (Y):

- house price

The model studies many examples of houses and their prices and learns the relationship between them.

Later, it can predict the price of a new house it has never seen before.

Supervised learning problems usually fall into two categories:

Regression  
Predicting continuous values (like price or temperature)

Classification  
Predicting categories (like spam vs not spam)

## Math Representation

Supervised learning tries to learn a function:

f(X) → Y

Where:

X = input features  
Y = output label

Example:

X = [house size, number of rooms]

Y = house price

The algorithm tries to learn the best function that maps inputs to outputs while minimizing prediction error.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(1)

# create simple regression dataset
X = np.linspace(0,10,50)
y = 3*X + 2 + np.random.normal(0,2,50)

plt.scatter(X,y)
plt.title('Simple Regression Dataset')
plt.xlabel("input feature (X) ")
plt.ylabel(" target / output feature (Y)" )
plt.show()
```

Explanation:

Each point represents a labeled training example.

The machine learning model learns the relationship between X and Y.

```python
from sklearn.model_selection import train_test_split

# reshape data
X_data = X.reshape(-1,1)

X_train, X_test, y_train, y_test = train_test_split(
    X_data, y, test_size=0.2, random_state=42
)

print("Training samples:",len(X_train))
print("Testing samples:",len(X_test))
```

Explanation:

We split the data into:

training set

testing set

Training data is used to teach the model.

Testing data evaluates how well the model performs.

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(X_train,y_train)

predictions = model.predict(X_test)

plt.scatter(X_test,y_test,label="True Values")
plt.scatter(X_test,predictions,label="Predictions")

plt.legend()
plt.title("Supervised Learning Prediction Example")

plt.show()
```

Explanation:

The model learns from training data and predicts outputs for new unseen data.
