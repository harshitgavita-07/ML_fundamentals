# K-Means Clustering

## Concept

K-Means is one of the most popular algorithms for **unsupervised learning**.

Unlike supervised learning, unsupervised learning does not use labeled data.

Instead, the algorithm tries to **discover hidden patterns or groups in the data**.

Clustering means grouping similar data points together.

Example:

Customer data

Features:
Age
Income
Spending score

K-Means can automatically group customers into clusters like:

Low spending customers
Medium spending customers
High spending customers

Pipeline:

Dataset
→ Choose number of clusters (K)
→ Initialize cluster centers
→ Assign points to nearest center
→ Update cluster centers
→ Repeat until convergence

# Math Behind K-Means

K-Means minimizes the **within-cluster variance**.

Objective function:

J = Σ || xᵢ − μⱼ ||²

Where:

xᵢ = data point  
μⱼ = cluster centroid  

Steps of the algorithm:

1. Choose K cluster centers randomly.

2. Assign each data point to the nearest centroid.

Distance metric (usually Euclidean):

distance = √((x1-x2)^2 + (y1-y2)^2)

3. Update centroid positions:

μ = mean of points assigned to the cluster

4. Repeat until centroids stop changing.

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs

X,y = make_blobs(
    n_samples=300,
    centers=3,
    random_state=42
)

plt.scatter(X[:,0],X[:,1])

plt.title("Dataset for Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")

plt.show()
```

This  dataset contains clusters that K-Means will discover.

```python
from sklearn.cluster import KMeans

model = KMeans(n_clusters=3)

model.fit(X)

labels = model.labels_

print(labels[:10])
```

The model assigns each point to a cluster label.

```python
plt.scatter(
    X[:,0],
    X[:,1],
    c=labels,
    cmap="viridis"
)

plt.scatter(
    model.cluster_centers_[:,0],
    model.cluster_centers_[:,1],
    color="red",
    marker="X",
    s=200
)

plt.title("K-Means Clustering Result")

plt.show()
```

Here you can see :

*   coloured clusters.
*   Red Markers are showing centriods.



# Summary

K-Means is an unsupervised learning algorithm used for clustering data.

Key ideas:

Clusters represent groups of similar data points.

Centroids represent the center of each cluster.

The algorithm iteratively improves cluster assignments.

Advantages:

Simple
Fast
Works well for many datasets

Limitations:

Requires choosing K
Sensitive to initialization
Assumes spherical clusters
