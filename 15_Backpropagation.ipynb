# Backpropagation

## Concept

Backpropagation is the algorithm that allows neural networks to learn.

When a neural network makes a prediction, it may produce an error.

Example:

Predicted value = 0.7  
Actual value = 1

Error = difference between prediction and true value.

Backpropagation works by sending this error **backwards through the network**
to update the weights.

Pipeline:

Input
→ Forward Pass
→ Prediction
→ Compute Loss
→ Backpropagate Error
→ Update Weights

This process is repeated many times until the model learns the correct patterns.

# Math Behind Backpropagation

Neural networks minimize a **loss function**.

Example (Mean Squared Error):

L = (y - ŷ)²

Where:

y = true value  
ŷ = predicted value

To update weights we compute the gradient:

∂L / ∂w

Using the **chain rule** from calculus:

∂L/∂w = (∂L/∂ŷ) * (∂ŷ/∂z) * (∂z/∂w)

Then we update weights using gradient descent:

w = w - α ∂L/∂w

Where:

α = learning rate

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-3,3,100)

y = x**2

plt.plot(x,y)

plt.title("Loss Function Example")
plt.xlabel("Weight Value")
plt.ylabel("Loss")

plt.grid(True)
plt.show()
```

This graph shows how gradient descent tries to reach the lowest loss.

```python
import numpy as np

# simple dataset
X = np.array([1,2,3,4])
y = np.array([2,4,6,8])

w = 0
lr = 0.01

for epoch in range(1000):

    y_pred = w * X

    loss = np.mean((y - y_pred)**2)

    grad = -2 * np.mean(X * (y - y_pred))

    w = w - lr * grad

print("Learned weight:", w)
```

this code demonstrates gradient-based learning .

```python
from sklearn.neural_network import MLPRegressor
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split

X,y = make_regression(n_samples=200,n_features=5)

X_train,X_test,y_train,y_test = train_test_split(
    X,y,test_size=0.2
)

model = MLPRegressor(
    hidden_layer_sizes=(20,10),
    max_iter=2000
)

model.fit(X_train,y_train)

score = model.score(X_test,y_test)

print("Model Score:",score)
```

This neural network uses backpropagation internally .

# Summary

Backpropagation is the algorithm that trains neural networks.

Key concepts:

Forward pass → prediction  
Loss function → measure error  
Backward pass → compute gradients  
Gradient descent → update weights

Without backpropagation, modern deep learning would not exist.

This algorithm powers:

Image recognition  
Speech recognition  
LLMs  
Autonomous systems
