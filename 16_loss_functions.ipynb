# Loss Functions

## Concept

A Loss Function measures how wrong a machine learning model's prediction is.

When a model makes a prediction, we compare it with the true value.

Example:

True value = 10  
Prediction = 8  

Loss = difference between them.

The goal of training is to **minimize the loss**.

Training pipeline:

Input Data  
→ Model Prediction  
→ Compute Loss  
→ Backpropagation  
→ Update Weights  

Different problems require different loss functions.

Common loss functions:

Mean Squared Error (MSE) → Regression problems  
Cross Entropy Loss → Classification problems

# Math Behind Loss Functions

1️⃣ Mean Squared Error (MSE)

Used in regression tasks.

Formula:

L = (1/n) Σ (y - ŷ)²

Where:

y = true value  
ŷ = predicted value  

This penalizes large prediction errors more strongly.

---

2️⃣ Binary Cross Entropy

Used for binary classification.

Formula:

L = - [y log(ŷ) + (1-y) log(1-ŷ)]

Where:

ŷ = predicted probability

---

3️⃣ Categorical Cross Entropy

Used for multi-class classification.

Formula:

L = - Σ yᵢ log(pᵢ)

This compares predicted probability distribution with the true distribution.

```python
import numpy as np
import matplotlib.pyplot as plt

y_true = 5
y_pred = np.linspace(0,10,100)

loss = (y_true - y_pred)**2

plt.plot(y_pred, loss)

plt.title("Mean Squared Error Loss Curve")
plt.xlabel("Predicted Value")
plt.ylabel("Loss")

plt.grid(True)
plt.show()
```

This graph shows how loss increases as predictions move away from the true Value.

```python
import numpy as np

def mse(y_true,y_pred):
    return np.mean((y_true - y_pred)**2)

def binary_cross_entropy(y_true,y_pred):
    return -np.mean(
        y_true*np.log(y_pred) + (1-y_true)*np.log(1-y_pred)
    )

y_true = np.array([1,0,1])
y_pred = np.array([0.9,0.2,0.8])

print("MSE:", mse(y_true,y_pred))
print("Binary Cross Entropy:", binary_cross_entropy(y_true,y_pred))
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

data = load_iris()

X_train,X_test,y_train,y_test = train_test_split(
    data.data,
    data.target,
    test_size=0.2,
    random_state=42
)

model = MLPClassifier(
    hidden_layer_sizes=(20,),
    max_iter=2000
)

model.fit(X_train,y_train)

accuracy = model.score(X_test,y_test)

print("Model Accuracy:", accuracy)
```

Internally this model uses cross-entropy loss.

# Summary

Loss functions measure how wrong a model's predictions are.

Training works by minimizing the loss.

Important loss functions:

Mean Squared Error → regression  
Binary Cross Entropy → binary classification  
Categorical Cross Entropy → multi-class classification

Loss functions guide gradient descent and backpropagation.

They are fundamental to training:

Neural Networks  
Deep Learning models  
Large Language Models
