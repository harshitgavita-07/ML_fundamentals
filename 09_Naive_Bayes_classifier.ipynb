# Naive Bayes Classifier

## Concept

Naive Bayes is a probabilistic machine learning algorithm used mainly for **classification problems**.

It is widely used in:

Spam detection  
Text classification  
Sentiment analysis  

The idea comes from **Bayes Theorem**, which calculates the probability of an event based on prior knowledge.

Example:

We want to classify an email as:

Spam  
Not Spam

The algorithm looks at the **words inside the email** and calculates the probability that the email belongs to each class.

Example words:

"free"
"offer"
"win"

If these words frequently appear in spam emails, the probability of the email being spam increases.

The algorithm is called **Naive** because it assumes that all features are **independent** of each other.

Example assumption:

The word "free" appearing in an email is independent of the word "offer".

This assumption is not always true, but surprisingly it works very well in practice.

Pipeline:

Features (words)
→ Probability calculation
→ Bayes theorem
→ Class prediction

# Math Behind Naive Bayes

The algorithm is based on **Bayes Theorem**:

P(A|B) = (P(B|A) * P(A)) / P(B)

Where:

P(A|B) = probability of class A given evidence B  
P(B|A) = probability of evidence B given class A  
P(A) = prior probability of class  
P(B) = probability of evidence  

For classification we compute:

P(Class | Features)

Example:

P(Spam | Words)

Using Bayes rule:

P(Spam | Words) ∝ P(Words | Spam) * P(Spam)

The class with the highest probability becomes the prediction.

Because of the independence assumption:

P(x1,x2,x3 | Class) =
P(x1|Class) * P(x2|Class) * P(x3|Class)

```python
import numpy as np
import matplotlib.pyplot as plt

classes = ['Spam', 'Not Spam']
probabilities = [0.8, 0.2]

plt.bar(classes, probabilities)
plt.title("Example Naive Bayes Prediction Probability")
plt.ylabel("Probability")
plt.show()
```

This graph shows how Naive Bayes chooses the class with the highest probability.

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

data = load_iris()

X_train, X_test, y_train, y_test = train_test_split(
    data.data,
    data.target,
    test_size=0.2,
    random_state=42
)

model = GaussianNB()

model.fit(X_train, y_train)

predictions = model.predict(X_test)

print("Predictions:", predictions[:10])
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, predictions)

print("Model Accuracy:", accuracy)
```

# Summary

Naive Bayes is a simple probabilistic classifier based on Bayes theorem.

Key ideas:

Uses probability to classify data.

Assumes features are independent (naive assumption).

Very fast and efficient.

Works extremely well for text-based problems.

Advantages:

Fast training
Works well with small datasets
Good for NLP tasks

Limitations:

Independence assumption is often unrealistic.
