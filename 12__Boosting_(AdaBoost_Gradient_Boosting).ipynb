# Boosting Algorithms

## Concept

Boosting is another ensemble learning technique.

Unlike Random Forest (which trains trees in parallel),
Boosting trains models **sequentially**.

Each new model focuses on correcting the mistakes made by the previous model.

Idea:

Weak learners → combine → strong learner.

Example:

Model 1 makes predictions but some points are wrong.

Model 2 focuses more on those wrong points.

Model 3 focuses on the remaining mistakes.

Final prediction is the **weighted combination** of all models.

Popular Boosting Algorithms:

AdaBoost  
Gradient Boosting  
XGBoost  
LightGBM  
CatBoost

Pipeline:

Dataset
→ Train weak model
→ Identify errors
→ Train next model focusing on errors
→ Combine models
→ Final prediction

# Math Behind Boosting

In AdaBoost we assign weights to each training example.

Initially:

All weights are equal.

After training a weak classifier:

Incorrect predictions get **higher weights**.

Correct predictions get **lower weights**.

This forces the next model to focus on harder samples.

Model weight:

α = ½ ln((1 − error) / error)

Final prediction:

F(x) = sign(Σ αᵢ hᵢ(x))

Where:

αᵢ = model weight  
hᵢ(x) = weak learner prediction

This creates a strong classifier by combining many weak learners.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification

X, y = make_classification(
    n_samples=200,
    n_features=2,
    n_redundant=0,
    n_clusters_per_class=1,
    random_state=42
)

plt.scatter(X[:,0], X[:,1], c=y, cmap="coolwarm")
plt.title("Dataset for Boosting")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

This shows a dataset where boosting can improve classification.

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.model_selection import train_test_split

from sklearn.datasets import load_iris

data = load_iris()

X_train, X_test, y_train, y_test = train_test_split(
    data.data,
    data.target,
    test_size=0.2,
    random_state=42
)

model = AdaBoostClassifier(
    n_estimators=50,
    random_state=42
)

model.fit(X_train, y_train)

predictions = model.predict(X_test)

print(predictions[:10])
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, predictions)

print("Boosting Accuracy:", accuracy)
```

Boosting often performs better than Random Forest on many datasets.

# Summary

Key concepts:

Weak learners

Error weighting
