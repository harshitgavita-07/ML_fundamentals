## Concept

Linear Regression tries to find the **best straight line** that describes the relationship between input and output data.

Example:

Input (X) → house size  
Output (Y) → house price

If house size increases, house price also increases.

The goal of linear regression is to find the **line that best fits the data**.

This line is called the **regression line**.

Equation of a straight line:

y = mx + b

Where:

x → input feature  
y → predicted value  
m → slope (how steep the line is)  
b → intercept (where the line crosses the y-axis)

## Math Behind Linear Regression

The hypothesis function is:

h(x) = θ0 + θ1x

Where:

θ0 = intercept  
θ1 = slope

The model tries to find θ0 and θ1 that minimize the prediction error.

We measure error using a **cost function**.

Cost Function:

J(θ) = (1 / 2m) Σ (h(x) - y)²

Where:

m = number of training examples

The goal is to minimize J(θ).

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

np.random.seed(42)
#Generate dataset
X= np.linspace(0,10,50)
Y = 4*X + 3 + np.random.normal(0,2,50)

plt.scatter(X,Y)
plt.title("Linear Regression Dataset ")
plt.xlabel(" Input features")
plt.ylabel("Target features")

plt.show()
```

Explanation:

Each point represents a real data example.

Our algorithm will try to draw the best line through these points

```python
from sklearn.linear_model import LinearRegression

X_Model = X.reshape(-1,1)

model = LinearRegression()

model.fit(X_Model,Y)
predictions = model.predict(X_Model)
```

Explanation:

The algorithm learns:slope

*   intercept
*   slope

that best describe the dataset.

```python
plt.plot(X,Y,label = "Real Data")
plt.plot(X,predictions,label = "Predicted Data")
plt.legend()
plt.title("Linear Regression Dataset ")
plt.xlabel("Input features")
plt.ylabel("Target features")
plt.show()
```

Explanation:

Blue dots = real data
Red line = prediction line

The algorithm tries to minimize error between points and the line.

```python
from sklearn.datasets import fetch_california_housing

data = fetch_california_housing()

X_real = data.data[:,0].reshape(-1,1)
y_real = data.target

model.fit(X_real,y_real)

pred_real = model.predict(X_real)

plt.scatter(X_real,y_real,alpha=0.2)

plt.plot(X_real,pred_real,color="red")

plt.title("Linear Regression on Real Housing Data")

plt.show()
```

Explanation:

Blue dots = real data
Red line = prediction line

The algorithm tries to minimize error between points and the line.
